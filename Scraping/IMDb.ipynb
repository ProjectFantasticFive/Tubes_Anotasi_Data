## Lanjutkan disini

```python
results = driver.find_elements_by_css_selector("div.lister-item.mode-detail")
```

```python
import urllib.request
```

```python
while (True):
    for result in results:
        driver.execute_script("return arguments[0].scrollIntoView();", result)
        #judul film
        judul = result.find_element_by_class_name("lister-item-header").find_element_by_tag_name('a').text

        #tahun film
        tahun = result.find_element_by_css_selector("span.lister-item-year.text-muted.unbold").text.split()
        if (len(tahun) == 2):
            tahun = tahun[1]
        else:
            tahun = tahun[0]

        #genre dan sinopsis singkat
        genre = result.find_element_by_class_name("genre").text.split(",")
        infolain = result.find_elements_by_tag_name('p')
        if (infolain[0].text == ''):
            infolain = [genre, infolain[2].text]
        else:
            infolain = [genre, infolain[1].text]

        #image url:
        image = result.find_element_by_css_selector("img.loadlate").get_attribute('src')
        list_of_films.append(
        {'judul': judul,
         'tahun': tahun,
         'genre': infolain[0],
         'sinopsis': infolain[1],
         'image': image,
        })
#         print(judul+"\n"+tahun+"\n"
#             +str(infolain[0])+"\n"+infolain[1]
#             +"\n"+image)
#         print("\n")
        
        #download image
        urllib.request.urlretrieve(image, "image/"+judul+".jpg")
        
    try:
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        next_button = driver.find_element_by_css_selector("a.flat-button.lister-page-next.next-page")
        next_button.click()
    except:
        break
    
    time.sleep(10)
    results = driver.find_elements_by_css_selector("div.lister-item.mode-detail")

    
```
